- 3.2) Who has the most field goals in 2020?

```python
# Boolean filtering by Year
is_2020=df['Year']==2020
print(is_2020.head())
```

```python
#Creating a variable for the Year
df_2020=df[is_2020]
print(df_2020.shape)
```

```python
field_goals_2020 = df_2020.groupby('Player').agg(Fields_Goal_2020 = ('Fields Goal', 'sum'))
field_goals_2020.sort_values(by='Fields_Goal_2020', ascending=False, na_position='last')
goals_15=field_goals_2020['Fields_Goal_2020']>=16
best_scorers=field_goals_2020[goals_15]
display(best_scorers.head(5))
```

```python
Players = ['Andre Drummond','Andrew Wiggins','D\'Angelo Russell','Jordan Clarkson','Marcus Morris']
Goals = best_scorers['Fields_Goal_2020']

# Figure Size
fig, ax = plt.subplots(figsize=(12, 5))
     
ax.barh(Players,Goals)
plt.title('Best Scorers of 2020')
plt.ylabel('Players')
plt.xlabel('Goals in 2020')
plt.show()
```

The top 3 players with most field goals in 2020 are: 
- Andrew Wiggins, 
- D'Angelo Russell
- Andre Drummond
